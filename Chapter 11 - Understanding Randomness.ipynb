## It's Not Easy Being Random

Humans tend to introduce non-randomness when attempting to generate "random" outcomes.

Numbers generated by a computer are **pseudo-random**: Not truly random in that they're generation is deterministic based on some starting point and a fixed algorithm.  They're often "good enough" when looking for a source of random numbers.

Some websites offer access to truly random numbers, typically generated using radioactive decay or other random physical phenomena.

* www.random.org
* www.randomnumbers.info

### Card Shuffling

* Research shows that using riffle shuffle, a deck isn't randomized until the shuffle has been performed **7 times**.  After seven, additional shuffles do little extra good.

## Simulation

* **Simulate**:  Immitate a real process so that we can manipulate and control the details of the process
* Each time we obtain a simulated answer is called a **trial**
* Each random selection within a trial that produces a part of the answer is a **component** of the simulation
* The trial's outcome is called the **response variable**

### Steps for Making a Simulation

* Specify how to model a component outcome using equally likely random digits
    * Identify the component to be repeated
    * Explain how to model the component's outcome
* Specify how to timulate trials
    * Explain how to combine the components to model a trial
    * State clearly what the response variable is
* Run the simulation
    * Run several trials
    * Create a chart to track results
* Analyze the response variable
    * Collect and summarize the results of all trials
    * State conclusions

## Step-by-Step Example: Simulation

* Plan: State the problem.  Identify the important parts of your simulation.
* Components: Identify the components.
* Outcomes: State how you will model each component using equally likely random digits.
* Trial: Explain how you will combine the components to simulate a trail.
* Response Variable: Define your response variable.

* Mechanics:  Run several trials.  Carefully record the random numbers, indicating:
    * the corresponding component outcomes
    * the value of the repsonse variable

* Analyze: Summarize the results across all trials to answer the initial question.
* Conclusion: Describe what the simulation shows, and interpret your results in the context of the real world.

## What Can Go Wrong?

* Don't overstate your case
    * Beware of confusing what _really_ happens with what a simulation suggests _might_ happen.
* Model outcome chances accurately
* Run enough trials
