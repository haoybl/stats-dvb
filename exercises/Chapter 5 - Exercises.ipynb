## 33

* us has a lower median and its central values are more skewed, it also has less variability in its central grouping of values
* other has more variability, but overall, a lower higher median mpg

```python
mpg = pd.read_table("../data/MPG.txt", names=['mileage', 'country'], skiprows=1)
mpg.head()
```

```python
# side-by-side histograms
us = mpg[mpg.country == 'US'].mileage
other = mpg[mpg.country != 'US'].mileage

bins = 10
rng = [15, 50]
ylim = [0, 7]

plt.subplot(1,2,1)
plt.suptitle("mpg - US vs. other")
plt.hist(us, bins=bins, range=rng)
plt.ylabel('count')
plt.xlabel('US')
plt.ylim(ylim)
plt.xlim(rng)
plt.subplot(1,2,2)
plt.hist(other, bins=bins, range=rng)
plt.ylim(ylim)
plt.xlim(rng)
plt.xlabel('other');
```

```python
plt.boxplot([us, other])

plt.xticks([1, 2], ['US', 'Other'])
plt.xlabel("Category")
plt.ylabel("MPG")
plt.title("MPG - US vs. Other");
```
